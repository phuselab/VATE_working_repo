```python
import pandas as pd

# Replace 'your_file.csv' with the path to your CSV file
df = pd.read_csv('/var/data/CMU_MOSI/label.csv')

df = df.sort_values(by=df.columns[0])
# Display the first few rows of the DataFrame
print(df)
```

```python
# Iterate over the DataFrame and create the text file for each row
root_path = "/var/data/CMU_MOSI/processed/"
enum = 0
for index, row in df.iterrows():
    # Create the filename using title1 and title2 columns
    filename = f"{row['video_id']}_{row['clip_id']}_{row['label']}.txt"
    
    # Write the content of the 'data' column into the text file
    with open(root_path + filename, 'w') as file:
        file.write(row['text'])

```

```python
import os
import shutil

# Define the main folder path where all the subject folders are
main_folder = '/var/data/CMU_MOSI/Raw'  # Update with your path
new_folder = '/var/data/CMU_MOSI/processed'    # Update with your destination path

# Loop through each subject folder in the main folder
for subject_id in os.listdir(main_folder):
    subject_folder = os.path.join(main_folder, subject_id)

    # Check if the item is a folder (subject folder)
    if os.path.isdir(subject_folder):
        # Loop through each .mp4 file in the subject folder
        for clip_file in os.listdir(subject_folder):
            if clip_file.endswith('.mp4'):
                # Extract clip_id by removing the file extension
                clip_id = os.path.splitext(clip_file)[0]

                # Create the new filename as 'subject_id_clip_id.mp4'
                new_filename = f'{subject_id}_{clip_id}.mp4'

                # Source file path
                old_file_path = os.path.join(subject_folder, clip_file)

                # Destination file path in the new folder
                new_file_path = os.path.join(new_folder, new_filename)

                # Copy and rename the file to the new folder
                shutil.copy2(old_file_path, new_file_path)

print("Files have been copied and renamed successfully!")
```

```python
import os
import glob

# Path to the folder containing the .txt and .mp4 files
folder_path = '/var/data/CMU_MOSI/processed'

# Get all the .txt files in the folder
txt_files = glob.glob(os.path.join(folder_path, '*.txt'))

for txt_file in txt_files:
    # Extract the base id and label from the .txt file name
    txt_filename = os.path.basename(txt_file)
    file_id = txt_filename.rsplit('_', 1)[0]

    label = os.path.splitext(txt_filename)[0].split('_')[-1]
    label = label.rsplit('.')[0]

    # Find the corresponding .mp4 file
    mp4_file = os.path.join(folder_path, f"{file_id}.mp4")
    
    if os.path.exists(mp4_file):
        # Construct the new mp4 file name with the label included
        new_mp4_filename = os.path.join(folder_path, f"{file_id}_{label}.mp4")
        
        # Rename the mp4 file
        os.rename(mp4_file, new_mp4_filename)
        print(f"Renamed {mp4_file} to {new_mp4_filename}")
    else:
        print(f"MP4 file for {file_id} not found.")
        os.remove(os.path.join(folder_path, txt_filename))
```

```python
import os
from collections import defaultdict

def rename_files_in_folder(folder_path):
    # Get a list of files in the folder
    files = os.listdir(folder_path)

    # Dictionary to track the progressive counters for each id_name
    file_counters = defaultdict(int)
    
    # Dictionary to store the assigned counters for each id_name
    assigned_counters = {}

    enum = 0

    for file_name in files:
        # Split the file name into 'id_name' and 'ext'
        id_name, ext = os.path.splitext(file_name)
        label = id_name.rsplit('_')[-1]
        take = id_name.rsplit('_')[-2]
        id_name = id_name.replace("_"+take+"_"+label, "")
        
        # If the id_name has not been assigned a counter, assign one
        if id_name not in assigned_counters:
            enum += 1
            file_counters[id_name] += enum
            assigned_counters[id_name] = file_counters[id_name]
        # print(assigned_counters)
        
        # Create the new file name with the assigned counter
        new_file_name = f"{assigned_counters[id_name]:2d}_{take}_{label}{ext}"  # Format counter with 3 digits
        
        # Get full file paths
        old_path = os.path.join(folder_path, file_name)
        new_path = os.path.join(folder_path, new_file_name)
        
        # Rename the file
        os.rename(old_path, new_path)

# Call the function with the folder path
folder_path = "/var/data/CMU_MOSI/processed/"  # Change this to your folder path
rename_files_in_folder(folder_path)
```
